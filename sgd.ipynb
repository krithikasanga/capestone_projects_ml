```python
import numpy as np
import pandas as pd
```

$
\hat{y} = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_n x_n
$

$
\frac{\partial L}{\partial \beta_0} = -2(y - \hat{y})
$

$
\beta_0 = \beta_0 - \eta \times \text{slope of intercept}
$

$
\frac{\partial L}{\partial \beta_m} = -2(y - \hat{y})x_m
$

$
\beta_m = \beta_m - \eta \times \text{slope of } \beta_m
$

```python
class Sgdregressor:
  def __init__(self,lr=0.01,epochs=100):
    self.coef_=None
    self.intercept_=None
    self.lr=lr
    self.epochs=epochs
  def fit(self,X_train,y_train):
    #initialize the coef and intercept
    self.coef_=np.ones(X_train.shape[1])
    self.intercept_=0

    for i in range(self.epochs):
      for j in range(X_train.shape[0]):
        #we are going to randomly select a row and based upon that row we are going to make 1 update on it
        #here there are 353 rows hence we run a loop and generate a random number and select that specific row to make an update it runs 353 times
        idx=np.random.randint(0,X_train.shape[0])

        y_hat=self.intercept_+np.dot(X_train[idx],self.coef_)

        intercept_der=-2*(y_train[idx]-y_hat)
        self.intercept_=self.intercept_ - (self.lr*intercept_der)

        coef_der = -2 * np.dot((y_train[idx] - y_hat),X_train[idx])
        self.coef_ = self.coef_ - (self.lr * coef_der)
    print(self.intercept_,self.coef_)

  def predict(self,X_test):
    return np.dot(X_test,self.coef_) + self.intercept_
```

```python
sgd = Sgdregressor(lr=0.01,epochs=40)
```

```python
start = time.time()
sgd.fit(X_train,y_train)
print("The time taken is",time.time() - start)
```

```python
y_pred = sgd.predict(X_test)
r2_score(y_test,y_pred)
```
