$$
\frac{\partial L}{\partial \beta_0}=-\frac{2}{n}\sum_{i=1}^{n}(y_i-\hat{y}_i)
$$

$$
\frac{\partial L}{\partial \beta_m}=-\frac{2}{n}\sum_{i=1}^{n}(y_i-\hat{y}_i)x_{im}
$$

```python
class Gdregressor:
  def __init__(self,epochs=100,lr=0.1):
    self.intercept_=None
    self.coef_=None
    self.lr=lr
    self.epochs=epochs

  def fit(self,X_train,y_train):
    #initialize the intercept to 0 and coefficients to 1
    self.intercept_=0
    self.coef_=np.ones(X_train.shape[1])#10 columns so 10 betas hence each beta is 1

    for i in range(self.epochs):
      #finding the intercept value #refer notebook for the formulas
      y_hat=self.intercept_+ np.dot(X_train,self.coef_)
      intercept_derv=-2*np.mean(y_train-y_hat)
      self.intercept_=self.intercept_-(self.lr * intercept_derv)

      #finding the coefficients
      coef_derv=-2*np.dot((y_train-y_hat),X_train)/X_train.shape[0]
      self.coef_=self.coef_-(self.lr*coef_derv)
    print(self.intercept_,self.coef_)

  def predict(self,X_test):
    return self.intercept_+np.dot(X_test,self.coef_)

```

```python
gdr=Gdregressor(1000,0.4)
```

```python
gdr.fit(X_train,y_train)
```

```python
y_pred=gdr.predict(X_test)
r2_score(y_test,y_pred)
```
